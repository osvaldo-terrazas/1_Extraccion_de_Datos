```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Filtro por objeto
filtro1 = df[df['NOMBRE_CLIENTE'] == 'PÚBLICO EN GENERAL']
filtro1
```

```python
#Filtro por filas
filtro2 = df.iloc[2:8, :]
filtro2
```

```python
#Filtro por columnas
filtro3 = df.iloc[ : , 2:8]
filtro3
```

```python
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col = 3)
df1
```

```python
#Filtro filas vs columnas
filtro4 = df1.loc[["SEVEN DAYS ALL SPORT", "REVUELTA MONTECARLO"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
#Filtro por cabeza
filtro5 = df.head(5)
filtro5
```

```python
#Filtro por cola
filtro6 = df.tail(5)
filtro6
```

```python
#Filtro por comparacion
filtro7 = df[df["PRECIO_UNITARIO"] < 1000]
filtro7
```

```python
#Filtro Y
filtro8 = df[(df["PRECIO_UNITARIO"] < 1000) & (df["FECHA_DOC"] == "2022-01-05")]
filtro8
```

```python
#Filtro o
filtro9 = df[(df["PRECIO_UNITARIO"] < 1000) | (df["FECHA_DOC"] == "2022-01-05")]
filtro9
```

```python
#Filtro not
filtro10 = df[(df["PRECIO_UNITARIO"] < 1000) & ~(df["FECHA_DOC"] == "2022-01-05")]
filtro10
```

```python
#Convertir filtros a archivos csv
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
filtro9.to_csv("Filtro9.csv")
filtro10.to_csv("Filtro10.csv")
```
